```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_principal = pd.read_excel('acoespuras.xlsx', sheet_name='Principal')
df_principal
```

```python
df_total_acoes = pd.read_excel('acoespuras.xlsx', sheet_name='Total_de_acoes')
df_total_acoes
```

```python
df_ticker = pd.read_excel('acoespuras.xlsx', sheet_name='Ticker')
df_ticker
```

Pegar as colunas 4 primeiras colunas do dataframe principal

```python
df_principal = df_principal[['Ativo','Data','Último (R$)','Var. Dia (%)']].copy()
df_principal
```

Renomear colunas

```python
df_principal = df_principal.rename(columns={'Último (R$)':'Valor_final', 'Var. Dia (%)':'Var_dia_pct'}).copy()
df_principal
```

```python
df_principal['Var_dia_pct'] = df_principal['Var_dia_pct'] / 100
df_principal
```

Criar uma nova coluna valor inicial

```python
df_principal['Valor_inicial'] = df_principal['Valor_final'] / (1 + df_principal['Var_dia_pct'])
df_principal
```

Criando merge de colunas

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

Drop da coluna Código

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

Renomear coluna Qtde. Teórica

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'Quantidade_acoes'}).copy()
df_principal
```

Criando coluna de Variação Real

```python
df_principal['Variacao_rn'] = (df_principal['Valor_final'] - df_principal['Valor_inicial']) * df_principal['Quantidade_acoes']
df_principal
```

Formatar numeros float tirar notação científica

```python
pd.options.display.float_format = '{:,.2f}'.format
```

```python
df_principal
```

Convertendo a colunas qunatidades de acoes para inteiro

```python
df_principal['Quantidade_acoes'] = df_principal['Quantidade_acoes'].astype(int)
df_principal
```

Criar a coluna resultado com os valores se subiu, desceu ou manteve a ação

```python
df_principal['Resultado'] = df_principal['Variacao_rn'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

Criar a coluna nome da ação

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

Criar novo dataframe para pegar o segmento e a idade das ações

```python
df_segmento_idade = pd.read_excel('Tabeladeacoes.xlsx', sheet_name='SegmentoIdade')
df_segmento_idade
```

```python
df_segmento_idade = df_segmento_idade.rename(columns={'Idade (em anos)':'Idade'}).copy()
df_segmento_idade
```

```python
df_principal = df_principal.merge(df_segmento_idade, left_on='Nome', right_on='Empresa', how='left')
df_principal = df_principal.drop(columns=['Empresa'])
df_principal
```

```python
# quero trocar o valor NAN por 20 na coluna Idade
df_principal['Idade'] = df_principal['Idade'].fillna(20)
```

```python
df_principal
```

```python
df_principal['Segmento'] = df_principal['Segmento'].fillna('Aviação')
df_principal
```

```python
df_principal['Idade'] = df_principal['Idade'].astype(int)
df_principal
```

```python
df_principal['Cat_Idade'] = df_principal['Idade'].apply(lambda x: 'Mais de 100' if x > 100 else ('Menos de 50' if x < 50 else 'Entre 50 e 100'))
df_principal
```

Pegando a maior, menor e a media das açoes que subiu e desceu

```python
maior = df_principal['Variacao_rn'].max()
menor = df_principal['Variacao_rn'].min()
media = df_principal['Variacao_rn'].mean()

media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rn'].mean()
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rn'].mean()

print(f'Maior\tR$: {maior:,.2f}')
print(f'Menor\tR$: {menor:,.2f}')
print(f'Média\tR$: {media:,.2f}')
print(f'Média de quem subiu\tR$: {media_subiu:,.2f}')
print(f'Média de quem desceu\tR$: {media_desceu:,.2f}')
```

As açoes que subiram

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

Somando o valores das açoes por segmento

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_rn'].sum().reset_index()
df_analise_segmento
```

Somando valores por resultado subui ou desceu

```python
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rn'].sum().reset_index()
df_analise_saldo
```

Criar gráfico de barras

```python
#fig = px.bar(df_analise_saldo, x='Resultado', y='Variacao_rn',text='Variacao_rn', title='Variação de Reais por Resultado')
#fig.show()
resultado = df_analise_saldo['Resultado']
variacao = df_analise_saldo['Variacao_rn']
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(resultado, variacao, color ='blue',edgecolor ='grey', 
        width = 0.4)
 
plt.xlabel("Resultado")
plt.ylabel("Variação de Reais")
plt.title("Variação de Reais por Resultado")
plt.show()
```
